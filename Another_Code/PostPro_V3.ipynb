```python
test['ATA'] = pd.to_datetime(test['ATA'])
def compute_post_processed_target_v6(row):
    key = (row['ARI_PO'], row['U_WIND'], row['V_WIND'], row['AIR_TEMPERATURE'], row['BN'])
    
    # CI_HOUR 값이 1 이하인 경우, 그대로 반환
    if row['CI_HOUR'] <= 1:
        return row['CI_HOUR']
    
    # 해당 샘플의 날짜 이후의 데이터 중, 키가 일치하고 CI_HOUR 값이 1 이하인 데이터만 필터링
    matching_rows = test[
        (test['ARI_PO'] == key[0]) & 
        (test['U_WIND'] == key[1]) & 
        (test['V_WIND'] == key[2]) & 
        (test['AIR_TEMPERATURE'] == key[3]) & 
        (test['BN'] == key[4]) & 
        (test['CI_HOUR'] <= 1) &
        (test['ATA'] > row['ATA'])
    ]
    
    # 일치하는 데이터가 없는 경우, 현재 값 반환
    if matching_rows.empty:
        return row['CI_HOUR']
    
    # 가장 가까운 일치하는 데이터의 날짜 선택
    nearest_matching_date = matching_rows['ATA'].min()
    
    # 현재 행의 날짜와 가장 가까운 일치하는 데이터의 날짜의 차이 계산
    time_difference = (nearest_matching_date - row['ATA']).total_seconds() / 3600
    
    return abs(time_difference)

# 함수 적용 및 결과 확인
test['Post_Processed_CI_HOUR_v6'] = test.apply(compute_post_processed_target_v6, axis=1)
test[['ATA', 'CI_HOUR', 'Post_Processed_CI_HOUR_v6']]
```

```python
import pandas as pd

# 1. Read the provided files
df1 = pd.read_csv('SOTATEST_TR_PO_CHANGE_CO_6to24_DISTPP.csv')
df3 = pd.read_csv('SOTATEST_V3_edit_1440to1680.csv')
corrected_df = pd.read_csv('corrected_test_with_target_v2.csv')

# 2. Calculate the difference between CI_HOUR values in the first two dataframes
merged_df_new = df1.merge(df3, on='SAMPLE_ID', suffixes=('_distpp', '_v3'))
merged_df_new['CI_HOUR_diff'] = (merged_df_new['CI_HOUR_distpp'] - merged_df_new['CI_HOUR_v3']).abs()

# 3. Extract the top 100 SAMPLE_IDs with the largest differences
top_100_diff_new = merged_df_new.sort_values(by='CI_HOUR_diff', ascending=False).head(100)
top_100_sample_ids = top_100_diff_new['SAMPLE_ID'].tolist()

# 4. Filter out the rows with the top 100 SAMPLE_IDs in the corrected dataframe
filtered_corrected_df = corrected_df[corrected_df['SAMPLE_ID'].isin(top_100_sample_ids)][['SAMPLE_ID', 'CI_HOUR']]

# 5. Create a comparison dataframe
comparison_df = top_100_diff_new.merge(filtered_corrected_df, on='SAMPLE_ID', how='left')
comparison_df = comparison_df[['SAMPLE_ID', 'CI_HOUR_distpp', 'CI_HOUR_v3', 'CI_HOUR', 'CI_HOUR_diff']]

# 6. Save the comparison dataframe to a new csv file
comparison_df.to_csv('comparison_output.csv', index=False)
```
